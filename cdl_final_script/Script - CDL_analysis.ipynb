```python
#attmeping to find the tournemanet averages for each team 
# Defineing the weights for each placement
weights = {
    'Tournament_Wins': 1,
    '_2nd_Place': 2,
    '_3rd_Place': 3,
    '_4th_Place': 4,
    '_5th_6th_Place': 5.5,
    '_7th_8th_Place': 7.5,
    '_9th_10th_Place': 9.5,
    '_11th_12th_Place': 11.5
}
```

```python
weighted_averages = []
```

```python
# Iterate over each row (each team)
for index, row in dataset_modified.iterrows():
    total_weighted_score = 0
    total_placements = 0
    
    # Calculate weighted scores and total placements
    for col, weight in weights.items():
        total_weighted_score += row[col] * weight
        total_placements += row[col]
    
    # Calculate weighted average placement
    weighted_avg = total_weighted_score / total_placements if total_placements > 0 else None
    weighted_averages.append(weighted_avg)

# Add the weighted averages to the dataset (if needed)
dataset_modified['Weighted_Average_Placement'] = weighted_averages
```

```python
#Filterin team names that have at least 1 tournament win
winning_teams = dataset_modified[dataset_modified['Tournament_Wins'] > 0]

#Grouping by Team_Name and adding up their tournament wins
tournament_wins = winning_teams.groupby('Team_Name')['Tournament_Wins'].sum()
tournament_wins.sort_values(ascending=False)
```

```python
# Define a function to display counts instead of percentages
def absolute_value(val):
    total = sum(tournament_wins)
    return f'{int(val/100*total)} wins'  # Convert percentage back to counts

#Creation of piechart
plt.figure(figsize=(8, 8))
plt.pie(tournament_wins, labels=tournament_wins.index, autopct=absolute_value, startangle=90, colors=plt.cm.Paired.colors)
plt.title('Tournament Wins Count (2021-2024)')
plt.show()
```

```python
#creation of piechart
plt.figure(figsize=(8, 8))
plt.pie(tournament_wins, labels=tournament_wins.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Tournament Wins (2021-2024)')
plt.show()
```

LAN VS ONLINE ENVIROMETNS 
- Understanding which teams historicaly perfromed in different enviroments 

```python
#avergaes for online series wins
average_onlinewin_perc = dataset_modified.groupby('Team_Name')['Stages_Online_Series_Win_Percentage'].mean()
sorted_online = average_onlinewin_perc.sort_values(ascending=False)
sorted_online
```

```python
#averages for major series wins 
average_majorwin_perc = dataset_modified.groupby('Team_Name')['Major_LAN_Series_Win_Percentage'].mean()
sorted_majors = average_majorwin_perc.sort_values(ascending=False)
sorted_majors
```

```python
# Merge both Series into a common DataFrame
win_percentages = pd.DataFrame({
    'Online': sorted_online,
    'LAN': sorted_majors
})

# Drop any rows (teams) where there might be missing values in either column
win_percentages.dropna(inplace=True)

# Sort the DataFrame based on one of the columns (e.g., Online win percentages)
win_percentages.sort_values(by='Online', ascending=False, inplace=True)

# The team names will automatically be the index of this DataFrame
win_percentages
```

```python
#understanding the correlation if you do well online then you will do well on LAN
core = win_percentages['Online'].corr(win_percentages['LAN'])
core
```

```python
# Set the bar width and positions for each team
bar_width = 0.35
index = np.arange(len(win_percentages))


plt.figure(figsize=(12, 8))

# Plot online win percentages
bars1 = plt.bar(index, win_percentages['Online'], bar_width, label='Online Win Percentage', color='skyblue')

# Plot LAN win percentages next to online win percentages
bars2 = plt.bar(index + bar_width, win_percentages['LAN'], bar_width, label='LAN Win Percentage', color='orange')


plt.xlabel('Teams')
plt.ylabel('Win Percentage (%)')
plt.title('Comparison of Online and LAN Win Percentages (2021-2024)')

# Adding team names to the x-axis (from the DataFrame's index)
plt.xticks(index + bar_width / 2, win_percentages.index, rotation=45, ha='right')


plt.legend()

# Add grid for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add text labels on the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.1f}%', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.1f}%', ha='center', va='bottom')

#plot
plt.tight_layout()
plt.show()
```

```python
def find_best_placing(row):
    
    placement_columns = {
        'Tournament_Wins': 1,
        '_2nd_Place': 2,
        '_3rd_Place': 3,
        '_4th_Place': 4,
        '_5th_6th_Place': 6,
        '_7th_8th_Place': 8,
        '_9th_10th_Place': 10,
        '_11th_12th_Place': 12
    }
    

    best_placing = float('inf')
    for col, placing in placement_columns.items():
        if row[col] > 0:
            best_placing = min(best_placing, placing)
    
    return best_placing


teams_placed_8th.loc[:, 'Best_Placing'] = teams_placed_8th.apply(find_best_placing, axis=1)
```
